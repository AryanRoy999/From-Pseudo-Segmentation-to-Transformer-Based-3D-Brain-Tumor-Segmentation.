```python
import os
import shutil
import random

# SOURCE: current dataset
SRC_DIR = r"C:\Users\aryan\OneDrive\Desktop\DLPROJECT\archive (3)-20260108T092958Z-1-001\archive (3)"

# DESTINATION: clean dataset
DST_DIR = r"C:\Users\aryan\OneDrive\Desktop\DLPROJECT\dataset"

# Split ratios
SPLITS = {
    "train": 0.7,
    "val": 0.15,
    "test": 0.15
}

random.seed(42)

for class_name in os.listdir(SRC_DIR):
    class_path = os.path.join(SRC_DIR, class_name)

    if not os.path.isdir(class_path):
        continue

    images = os.listdir(class_path)
    images = [img for img in images if img.lower().endswith(('.png', '.jpg', '.jpeg'))]

    random.shuffle(images)

    n = len(images)
    train_end = int(n * SPLITS["train"])
    val_end = train_end + int(n * SPLITS["val"])

    split_files = {
        "train": images[:train_end],
        "val": images[train_end:val_end],
        "test": images[val_end:]
    }

    for split, files in split_files.items():
        out_dir = os.path.join(DST_DIR, split, class_name)
        os.makedirs(out_dir, exist_ok=True)

        for file in files:
            src_file = os.path.join(class_path, file)
            dst_file = os.path.join(out_dir, file)
            shutil.copy(src_file, dst_file)

print(" Dataset successfully split into train / val / test")
```
